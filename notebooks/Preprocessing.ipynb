Next, we will keep only those records for which we have the non-null value of the response variable i.e. ALF. 

```python
data_train = df[(df['ALF'].notnull())]
```

Let's separate the continuous and categorical variables for better understanding. 

```python
continuous_cols = ['Age', 'Weight','Height','Body Mass Index', 'Waist', 'Maximum Blood Pressure', 'Minimum Blood Pressure', 
                   'Good Cholesterol','Bad Cholesterol', 'Total Cholesterol']
categorical_cols = list(set(df.columns)-set(continuous_cols))
```

We will now replace the empty values in the continuous features with the mean of that particular feature column.

```python
df.head()
for col in continuous_cols:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
# All continuous columns without the age
continuous_cols_1 = ['Weight','Height','Body Mass Index', 'Waist', 'Maximum Blood Pressure', 'Minimum Blood Pressure', 
                   'Good Cholesterol','Bad Cholesterol', 'Total Cholesterol']
```

We will calculate the normality of the continuous features by checking the skewness and kurtosis values.

```python
for col in continuous_cols_1:
    print(col + ': Kurt -->',"{0:.4f}".format(df[col].kurt()))
```

```python
for col in continuous_cols_1:
    print(col + ': Skewness --> ',"{0:.4f}".format(df[col].skew()))
```

```python
#PCA Analysis
#OneHotEncoding
#Scale Data
```

It has been found that the features are a little skewed and to remove the skewness and make them normal we will apply the box-cox transformation. <br><br>
Also, we have some correlated features such as BMI and Total Cholesterol. We will get rid of these features as well.

```python
continuous_cols_2 = [ 'Age', 'Weight','Height','Waist', 'Maximum Blood Pressure', 'Minimum Blood Pressure', 
                   'Good Cholesterol','Bad Cholesterol']

for col in continuous_cols_2:
    nd = np.asarray(df[col].values)
    dft = stats.boxcox(nd)[0]
    df[col] = dft.tolist()

df.drop(columns = ['Body Mass Index', 'Total Cholesterol'], inplace=True)
df.head()
```

For fetching accurate results through our model, we need to scale the feature data.

```python
# Scaling
df[continuous_cols_2] = StandardScaler().fit_transform(df[continuous_cols_2])
df.head()
```

```python
df[continuous_cols_2].skew()
```

```python
df[continuous_cols_2].kurt()
```

Now, let's perform some analysis on the catergorical features. Firstly, we will convert the categories into the nominal data as integers and for each missing value we will create a new category with value as -1.

```python
#Categorical

name_category = ['Gender', 'Region', 'Source of Care']

df.loc[(df.Gender == 'M'),'Gender']=1
df.loc[(df.Gender == 'F'),'Gender']=0
#0=F, 1=M

df.loc[(df.Region == 'east'),'Region']=1
df.loc[(df.Region == 'west'),'Region']=0
df.loc[(df.Region == 'north'),'Region']=2
df.loc[(df.Region == 'south'),'Region']=3

#0=West, 1=East, 2=North, 3=South


df.loc[(df['Source of Care'] == 'Governament Hospital'),'Source of Care']=1
df.loc[(df['Source of Care'] == 'Private Hospital'),'Source of Care']=0
df.loc[(df['Source of Care'] == 'clinic'),'Source of Care']=2
df.loc[(df['Source of Care'] == 'Never Counsulted'),'Source of Care']=3


df['Source of Care']
```
